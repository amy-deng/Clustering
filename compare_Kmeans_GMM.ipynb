```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture

f = plt.figure(figsize=(12, 12))
n_samples = 500
random_state = 300
k = 4

X_varied, y_varied = make_blobs(n_samples=n_samples,
#                                 cluster_std=[1.0, 2.5, 0.5, .5, .5],  # Different variance
                                random_state=random_state) 
kmeans = KMeans(n_clusters=k, random_state=random_state)
y_pred = kmeans.fit_predict(X_varied)
M1 = kmeans.cluster_centers_
print 'centers of k-means\n',M1
plt.subplot(221)
plt.scatter(X_varied[:, 0], X_varied[:, 1], 10, c=y_pred)
plt.scatter(M1[:, 0], M1[:, 1], 70, marker='x',color='red')
plt.title("k-means")


gmm = GaussianMixture(n_components=k, covariance_type='full', max_iter=120, random_state=0)
gmm.fit(X_varied)
y_pred = gmm.predict(X_varied)
M2 = gmm.means_
print 'means of GMM\n',M2
plt.subplot(222)
plt.scatter(X_varied[:, 0], X_varied[:, 1], 10,c=y_pred)
plt.scatter(M2[:, 0], M2[:, 1], 70, marker='x',color='red')
plt.title("Gaussian Mixture")

plt.show()
# f.savefig("kmeans.pdf")
```
